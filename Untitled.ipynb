```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pyautogui
import os
import time
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains


# ⇩ Variaveis 
navegador = webdriver.Chrome() # Abri Navegador Google Chrome
navegador.maximize_window()
planilha = pd.read_excel('Nomes.xlsx')

#Tempo de Espera
time.sleep(3)

#Alerta 1 
pyautogui.confirm(text='DEIXE O NAVEGADOR GOOGLE CHROME ABERTO PARA EVITAR ERROS', title='ROBÔ CREFISA', buttons=['OK'])
pyautogui.confirm(text='VERIFIQUE SE TODOS OS NOMES ESTÃO CORRETOS,NÃO ESTÃO DUPLICADOS OU COM STATUS JÁ INSERIDO', title='ROBÔ CREFISA', buttons=['OK'])

#Abrindo Pasta Procon CREFISA
navegador.get('')

time.sleep(5) #Tempo de Espera

#Login
navegador.find_element(By.XPATH,'//*[@id="i0116"]').send_keys("")
navegador.find_element(By.XPATH,'//*[@id="idSIButton9"]').send_keys(Keys.ENTER)

time.sleep(2) #Tempo de Espera

#Login
navegador.find_element(By.XPATH,'//*[@id="i0118"]').send_keys("")
navegador.find_element(By.XPATH,'//*[@id="idSIButton9"]').send_keys(Keys.ENTER)

time.sleep(2) #Tempo de Espera                       

navegador.find_element(By.XPATH,'//*[@id="idBtn_Back"]').send_keys(Keys.ENTER) #Não Manter Conectado

time.sleep(10) #Tempo de Espera


#Modo de exibiçao de grade
navegador.find_element(By.XPATH,'//*[@id="appRoot"]/div[1]/div[2]/div[3]/div/div[2]/div[2]/div[2]/div[1]/div/div/div/div/div/div/div[1]/div[3]/button').click()
time.sleep(3)

#Loop até o termino dos nomes na planilha
for i, nome in enumerate (planilha["Nomes"]): # Variavel "nome" guarda os valores dos nomes, enumera os nomes.

    navegador.find_element(By.XPATH,'//*[@id="sbcId"]/form/input').click() #selecionando campo de busca
    navegador.find_element(By.XPATH,'//*[@id="sbcId"]/form/input').send_keys(Keys.CONTROL + 'a', Keys.BACKSPACE) # Limpando campo de busca
    time.sleep(2) # Tempo de espera
    navegador.find_element(By.XPATH,'//*[@id="sbcId"]/form/input').send_keys(nome + Keys.ENTER) #Inserido nomes e efetuando a pesquisa
    time.sleep(5) # Tempo de espera
    navegador.find_element(By.CSS_SELECTOR,'tr.gridRow_d8fb0729:nth-child(2) > td:nth-child(8) > div:nth-child(1)').click() #Clicks para abrir os Status
    navegador.find_element(By.CSS_SELECTOR,'tr.gridRow_d8fb0729:nth-child(2) > td:nth-child(8) > div:nth-child(1)').click() #Click para abrir os Status
    time.sleep(3) # Tempo de espera
    navegador.find_element(By.XPATH,'//*[@id="Flyout-1"]').click()
    time.sleep(2.0) # Tempo de espera
    navegador.find_element(By.CSS_SELECTOR,'tr.gridRow_d8fb0729:nth-child(3) > td:nth-child(8)').click()
    time.sleep(1.5) # Tempo de espera

    
pyautogui.confirm(text='CLIQUE EM OK PARA FECHAR O NAVEGADOR', title='ROBÔ CREFISA', buttons=['OK']) #Alerta 2 
navegador.quit() # Fechar navegador
    
    
```
